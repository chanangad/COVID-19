# function to read json and output text from json

```python
def data_reader(fname):
    with open(fname) as f:
        data = json.load(f)
    
    try:
        abstract = data['abstract'][0]['text'] + '\n\n'
    except:
        abstract = ''
    
    body = ''
    for item in data['body_text']:
        body += item['text']+'\n\n'
        
    text = abstract + body
    
    return data['paper_id'], text
```

```python
paper_id_list = []
paper_body_list = []

for i,fname in enumerate(flist):
    print(i+1, " : ", fname)
    paper_id, text = data_reader(folder_name+fname)
    paper_id_list.append(paper_id)
    paper_body_list.append(text)
    
```

# combine into dataframe and export to csv

```python
med_df = pd.DataFrame(list(zip(paper_id_list, paper_body_list)), columns=['paper_id', 'body'])
```

```python
med_df['body'] = med_df['body'].astype('str')
med_df['paper_id'] = med_df['paper_id'].astype('str')
```
